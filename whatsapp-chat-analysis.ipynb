<b>Some fun with WhatsApp chat data</b>

<br>

<p><b>How to get data?</b></p>
Steps:

1. Open any WhatsApp chat on your mobile [Andriod]

2. Goto options [right top]-> more -> Export Chat

Select the non-media option, since we are only interested in text messages.

This will download all the chats from beginning as a text file with filename as Group Name or the name of the person.

The format of the file is -

<b><i>Date, Time - Name: Message</i></b>

eg.

5/7/21, 2:44 AM - Peter Parker: Hello


There are few corner cases that does not follow the above pattern.. These are the WhatsApp generated messages -
few eg. 

Encryption Related -  

5/7/21, 2:44 AM - Messages and calls are end-to-end encrypted. No one outside of this chat, not even WhatsApp, can read or listen to them. Tap to learn more.


Group Created Message - 

5/7/21, 2:44 AM - Peter Parker created group "New Group"


Group Name Change -

5/7/21, 1:36 PM - You changed the subject from "New Group" to "Yet Another Group"


Similarly for - adding member, changing the icon, someone left, you are blocked, etc... 


<br>
<br>
Also one important thing to notice is that, the chat can be multiline. 

For this fun activity, i will be exploring WhatsApp chat from one of my group

```python
file_name = "group_chat.txt"
```

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = []
with open(file_name) as fp:
    for line in fp:
        if not (' - ' in line and ': ' in line):
            continue
        try:
            name = line.split(' - ')[1].split(': ')[0]
            if len(name) == 0 or ":" in name:
                continue
            
            date = line.split(',')[0]
            data.append((date, name))
        except Exception as e:
            print(line, e)
```

```python
df = pd.DataFrame(data, columns = ["Date", "Name"])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
output = []
for d in df[df["Date"]>"2022-01-01"].groupby('Date'):
    date = d[0]
    temp = {}
    temp["Date"] = date
    for i in d[1].groupby('Name'):
        temp[i[0]] = len(i[1])        
    output.append(temp)
```

```python
fdf = pd.DataFrame(output)
```

```python
fdf
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Bar(name=name, x=fdf['Date'], y=fdf[name]) for name in df["Name"].unique()
])
# Change the bar mode
fig.update_layout(barmode='stack')
fig.show()
```
